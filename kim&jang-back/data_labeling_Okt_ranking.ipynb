```python
import pandas as pd
f = pd.read_csv('korean_hate_speech.csv')
f.tail(10)
```

```python
idx = []
idx.extend(f[f['contain_gender_bias']==True].index)
print(idx[:10])
```

```python
idx2 = []
idx2.extend(f[(f['bias']=='gender')|(f['bias']=='others')].index)
print(idx2[:10])
```

```python
idx3 = []
idx3.extend(f[(f['hate']=='hate')|(f['hate']=='offensive')].index)
print(idx3[:10])
```

```python
index = idx + idx2 + idx3
y = sorted(index)

f_idx = []
for x in y:
    if x not in f_idx:
        f_idx.append(x)
print(f_idx)
```

```python
f['label'] = '0'
f.head()
```

```python
#부정 1, 긍정 0
f.loc[f_idx,'label'] = '1'
f.head()
#f.to_csv('hate_speech_labeled.csv', encoding = 'utf-8')
```

```python
txt = f['comments']
```

```python
#특수문자 제거
import re
compile = re.compile("[^ ㄱ-ㅣ가-힣]+")
for i in range(len(txt)):
    a = compile.sub("",txt[i])
    txt[i] = a
print(txt)
```

```python
from konlpy.tag import Okt
okt = Okt()
result =[]
result = [okt.nouns(i) for i in txt]
ft= [r for i in result for r in i]
print(ft)
```
